```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://alongat:123456@localhost:5432/israeli_prices', client_encoding='utf8')
```

### Load csvs and upload to postgresql

```python
# the csvs were created in the ParseXMLs notebook..
# load csvs
allPrices = pd.concat([pd.read_csv('../csv/CoopShopPrices_2017_06_29.csv'),
                       pd.read_csv('../csv/ShupersalPrices_2017_06_29.csv'),
                       pd.read_csv('../csv/tivTaamPrices_2017_06_29.csv')])
```

```python
# create a table 'prices' on postgresql with the data
allPrices.to_sql('prices', engine)
```

### Run some queries

```python
pd.read_sql_query("select * from prices where chain='קואופ שופ' limit 10", engine)
```

```python
q = """
select "ItemName", "ItemPrice", "chain", "branch"
from prices
where "ItemName" like '%%קולה%%'
order by "ItemPrice"
limit 20
"""

pd.read_sql_query(q, engine)
```
